```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pymongo
from pprint import pprint
from livestockwatch.manualdata_utils import DataProvider
```

```python
cs_path = "all_cs.xlsx"
if not os.path.exists(cs_path):
    base_path = "/Volumes/fitramhd/BISE/WA02_month/"
    lst_excel_path = [
        os.path.join(base_path, "2015-7-data.xlsx"),
        os.path.join(base_path, "2015-6-data.xlsx"),
        os.path.join(base_path, "manual_data.xlsx")
    ]

    lst_cs_df = []

    data_provider = DataProvider(excel_result_path=cs_path)

    for excel_path in lst_excel_path:
        _, cs_df = data_provider.process_excel_ca(excel_path)
        if not cs_df.empty:
            lst_cs_df.append(cs_df)
            
    if lst_cs_df:
        cs_df_writer = pd.ExcelWriter(cs_path)
        cs_df = pd.concat(lst_cs_df).reset_index()
        cs_df = cs_df.drop("index", 1)
        cs_df.sort_values(by=["datetaken", "cs_name", "frametime"], inplace=True)
        cs_df.to_excel(cs_df_writer, index=False, sheet_name="all")
        cs_df_writer.save()
else:
    excel_handler = pd.ExcelFile(cs_path)
    excel_data = pd.read_excel(excel_handler, sheetname=None)
    cs_df = excel_data["all"]
```
