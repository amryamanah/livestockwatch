```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.charts import Bar
output_notebook()
```

```python
excel_handler = pd.ExcelFile("april172016.xlsx")
excel_data = pd.read_excel(excel_handler, sheetname=None)
```

```python
plr_df = excel_data["plr_all"]
plr_df = plr_df[plr_df.same_region]
plr_df.region.groupby(plr_df.region).count().plot(kind='bar')
```

```python
plr_df.region.groupby(plr_df.region).count()
```

## All Region

```python
low_vita = plr_df[plr_df.vit_a <= 40]
print(low_vita.shape)
low_vita.region.groupby(low_vita.region).count().plot(kind="bar")
low_vita.region.groupby(low_vita.region).count()
```

```python
high_vita = plr_df[plr_df.vit_a >= 60]
print(high_vita.shape)
high_vita.region.groupby(high_vita.region).count().plot(kind="bar")
high_vita.region.groupby(high_vita.region).count()
```

```python
high_vita.svg_ca.plot.box()
```

```python
low_vita.svg_ca.plot.box()
```

```python
high_vita.svg_ca.mean()
```

```python
low_vita.svg_ca.mean()
```

```python
p = figure(width=800, height=500,  x_axis_label = "Vit A (IU/DL)",
       y_axis_label = "CA")
p.circle(low_vita.vit_a,low_vita.svg_ca, size=10, color="red", alpha=0.5, legend="Low vitamin A < 35")
p.triangle(high_vita.vit_a,high_vita.svg_ca, size=10, color="blue", alpha=0.5, legend="High vitamin A > 60")
show(p)
```
